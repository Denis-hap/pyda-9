## Задание 1

```python

from scipy.stats import norm
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
znachen = [
    [-1, 5],
    [15, 10],
    [55, 102],
    [-5, 10],
    [1, 8]
]
```

```python
for vals in znachen:
    norm_rv = sts.norm(loc=vals[0], scale=vals[1])
    x = np.linspace(-5, 14, 100)
    y = norm_rv.pdf(x)
    plt.plot(x, y, label=f"µ = {vals[0]}, σ = {vals[1]}")
```

```python
from scipy.stats import expon
```

```python
for vals in znachen:
    expon_rv = expon(loc=vals[0], scale=vals[1])
    x = np.linspace(-5, 14, 100)
    y = expon_rv.pdf(x)
    plt.plot(x, y, label=f"µ = {vals[0]}, σ = {vals[1]}")
```

```python
from scipy.stats import t
```

```python
for vals in znachen:
    t_rv = t(loc=vals[0], scale=vals[1], df = 1)
    x = np.linspace(-5, 14, 100)
    y = t_rv.pdf(x)
    plt.plot(x, y, label=f"µ = {vals[0]}, σ = {vals[1]}")
```
