```python
remove_outliers = df_8counts 

def passes(line_):
    
    q1 = remove_outliers[line_].quantile(0.25)
    q3 = remove_outliers[line_].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    
    print(remove_outliers[line_][remove_outliers[line_].between(lower_bound, upper_bound, inclusive=False)].describe())
    
    return (remove_outliers[line_][remove_outliers[line_].between(lower_bound, upper_bound, inclusive=False)].count(), 
            remove_outliers[line_].count() -
            remove_outliers[line_][remove_outliers[line_].between(lower_bound, upper_bound, inclusive=False)].count() )


for i in df_8counts:
    print(f'{passes(i)} (было строк, исключили строки)')
    

```
