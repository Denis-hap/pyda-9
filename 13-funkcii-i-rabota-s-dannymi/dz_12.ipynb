```python
data_estimation = pd.read_csv('ml-100k/u.data', names = ['Пользователь','Фильм', 'Оценка','Чго-то еще']  , sep = '\t')
data_estimation.head()
```

```python
def group_rating(param):
    
    if param <= 2:
        result = 'Низкий рейтинг'
        
    if (param >= 3) and (param < 4.5):
        result = 'Средний рейтинг'
        
    if (param >= 4.5) and (param <= 5):
        result = 'Высокий рейтинг'
    
    return result
```

```python
data_estimation['gr_rating'] = data_estimation['Оценка'].apply(group_rating)
```

```python
data_keywords = pd.read_csv('keywords.csv') 
data_keywords.head()
```

```python
def geo_data(data_keyword):
    
    result = 'undefined'
    
    if (('москва') or ('тула') or ('ярославль')) in data_keyword.split(' '):
        result = 'Центр'

    if (('петербург') or ('псков') or ('мурманск')) in data_keyword.split(' '):
        result = 'Северо-Запад'

    if (('владивосток') or ('сахалин') or ('хабаровск')) in data_keyword.split(' '):
        result = 'Дальний Восток'        
        
        
    return result
```

```python
data_keywords['region'] = data_keywords['keyword'].apply(geo_data)
data_keywords[data_keywords['region'].isin(['Центр', 'Северо-Запад', 'Дальний Восток'])]

```

Задача 3


```python
sequence = range(1950,2011,1)

years = [num for num in sequence]
```

```python
data_movies = pd.read_csv('ml-latest-small/movies.csv')
data_movies.head()
```

```python
data_ratings = pd.read_csv('ml-latest-small/ratings.csv')
data_ratings.head()
```

```python
def production_year(production_year, years):
    
    if re.findall(r'\d{4}', production_year):
        result_year = re.findall(r'\d{4}', production_year)
        
#Проверка года из списка дат
        if int(result_year[0]) in years:
            return int(result_year[0])

    return 1900
        
```

```python
data_movies['year'] = data_movies['title'].apply(production_year, years = years ) 

data_movies
```

```python
mean_data_ratings = data_ratings.groupby('movieId').mean()
data_ratings.groupby('movieId').mean()
```

```python
data_movies_all = data_movies.join(mean_data_ratings)

data_movies_all_year = data_movies_all.groupby('year').mean()['rating']

data_movies_all_year.sort_values(ascending = False)
```
